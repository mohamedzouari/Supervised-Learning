# Detecting Colinearity with Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(df.corr())
```

```python
pd.Series(eigenvalues).sort_values()
```

smaller values represent presence of colinearity

```python
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending=False)
```

```python
print(df.columns[2], df.columns[8], df.columns[9])
```

these are the factors that are causing the multicollinearity

 # Revising Feature importance and extractions


Check:

   1. Direction of the coefficient
   2. Impact of the variable/coefficient on the model

## Method1: Standarize variable to identify key features

In order to perform 2. properly, we need to standarise the variable

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y);
```

```python
result = pd.DataFrame(list(zip(model.coef_,df.columns)), columns=['coefficient','name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
standard_coef_linear_reg = make_pipeline(scaler, model)
```

```python
standard_coef_linear_reg.fit(X,y)
result = pd.DataFrame(list(zip(standard_coef_linear_reg.steps[1][1].coef_,df.columns)),
                      columns=['coefficient','name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

## Method2: Use $R^2$ to identify key features

* Compare $R^2$ of model against $R^2$ of model without a feature. 

* A significant change in $R^2$ signify the importance of the feature.



```python
from sklearn.metrics import r2_score
```

```python
linear_reg = smf.ols(formula = 'y ~ LSTAT + DIS + RM + RAD + TAX + PTRATIO + NOX + ZN + CRIM + B + CHAS + INDUS + AGE',
                    data=df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
linear_reg = smf.ols(formula = 'y ~  DIS + RM + RAD + TAX + PTRATIO + NOX + ZN + CRIM + B + CHAS + INDUS + AGE',
                    data=df)
benchmark_NO_LSTAT = linear_reg.fit()
r2_score(y, benchmark_NO_LSTAT.predict(df))
```

```python
linear_reg = smf.ols(formula = 'y ~ LSTAT + DIS + RM + RAD + TAX + PTRATIO + NOX + ZN + CRIM + B + CHAS + INDUS ',
                    data=df)
benchmark_NO_AGE = linear_reg.fit()
r2_score(y, benchmark_NO_AGE.predict(df))
```
