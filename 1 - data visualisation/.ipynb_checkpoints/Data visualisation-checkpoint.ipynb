```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="ticks", color_codes=True)
%matplotlib inline
```

```python
adultdata = pd.read_csv('adult.data',header=-1)
```

```python
col_names = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']
adultdata.columns = col_names
adultdatahead = adultdata.head()
adultdata.head()
```

```python
adultdata.groupby('salary').size()
```

## Visualisation

### seaborn.pairplot

https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(adultdata,hue='sex', x_vars=["age"],
...                  y_vars=["education-num"],size=5,kind="reg");
```

### seaborn.violonplot

https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
plt.figure(figsize=(35,22))
plt.subplot(2,2,1)
sns.violinplot(y='age',x='education',data=adultdata,hue='sex');
```

### matplotlib.pyplot.hist

```python
adultdatahead.hist(edgecolor='black',linewidth=1.2, figsize=(12,8))
plt.show()
```
